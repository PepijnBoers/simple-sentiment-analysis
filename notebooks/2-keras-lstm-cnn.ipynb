```python
from src.models.train_model_keras import load_data
from sklearn.metrics import confusion_matrix
from src.models.predict_model_keras import make_prediction, load_pickle
from src.visualization.visualize import plot_confusion_matrix, show_mistakes
from src.data.utils import find_mistakes
from tensorflow import keras
```

```python
DATA_PATH = "../data/processed"

TRAIN_PATH = f"{DATA_PATH}/train250.csv"
TEST_PATH = f"{DATA_PATH}/test50.csv"

TOKENIZER_PATH = "../models/tokenizer.pickle"

LSTM_MODEL_PATH = "../models/keras_20210626-230333_5000_200_lstm.h5"
CNN_MODEL_PATH = "../models/keras_20210626-232734_5000_200_cnn.h5"
```

```python
# Load data.
X_train, y_train = load_data(TRAIN_PATH)
X_test, y_test = load_data(TEST_PATH)

# Load model.
model_lstm = keras.models.load_model(LSTM_MODEL_PATH)
model_cnn = keras.models.load_model(CNN_MODEL_PATH)

# Load tokenizer.
tokenizer = load_pickle(TOKENIZER_PATH)
```

```python
# Make predictions lstm
y_pred = make_prediction(model_lstm, X_test, tokenizer, 200)

# PLot confusion matrix lstm
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
plot_confusion_matrix(cm, "Confusion Matrix LSTM", ['Negative', 'Positive'], save=True)
```

```python
# Make predictions cnn
y_pred = make_prediction(model_cnn, X_test, tokenizer, 200)

# PLot confusion matrix cnn
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
plot_confusion_matrix(cm, "Confusion Matrix CNN", ['Negative', 'Positive'], save=True)
```

```python
# Show some examples that were predicted incorrectly.
indices = find_mistakes(y_test.to_numpy(), y_pred)
show_mistakes(X_test, indices, y_pred, 5)
```
