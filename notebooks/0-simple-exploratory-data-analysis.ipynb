```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
DATA_PATH = "../data/interim/train250.csv"
```

```python
# Load data
data_df = pd.read_csv(DATA_PATH)
X_train = data_df['body']
y_train = data_df['label']
```

```python
# Distribution review length.
plt.hist([len(review) for review in X_train.to_list()])
plt.title("Review length distribution")
plt.show()
```

```python
# Distribution positive and negative reviews.
labels = ['Negative', 'Positive']
negative_reviews, positive_reviews = np.bincount(y_train.to_list())
plt.pie([positive_reviews, negative_reviews], labels=labels, 
        autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2, radius=1.5)
plt.suptitle('Label Distribution', size=16, y=1.12)
plt.show()
```

```python
# Example of positive and negative review.
positive_reviews = np.where(y_train == 1)
negative_reviews = np.where(y_train == 0)

print(f"Positive review:{X_train.iloc[positive_reviews[0][0]]}\n")
print(f"Negaitve review:{X_train.iloc[negative_reviews[0][0]]}")
```
